```python
def avgMPchange(T):
    newArray = []
    for i in range(N):
        # 前面大部分元素
        if i<=N-1-T:
            tmp = (np.mean(midprice[i:i+T])-midprice[i])/midprice[i]
            newArray.append(tmp)
        # 对于最后一个元素 由于其没有后继 故补充0
        elif i==N-1:
            newArray.append(0)
        # 尾部元素只需要计算剩下的全部值的均值即可
        else:
            tmp = (np.mean(midprice[i:-1])-midprice[i])/midprice[i]
            newArray.append(tmp)
    return newArray
```

# 封装

```python
def avgMPchange(T,N,midprice):
    newArray = []
    for i in range(N):
        # 前面大部分元素
        if i<=N-1-T:
            tmp = (np.mean(midprice[i:i+T])-midprice[i])/midprice[i]
            newArray.append(tmp)
        # 对于最后一个元素 由于其没有后继 故补充0
        elif i==N-1:
            newArray.append(0)
        # 尾部元素只需要计算剩下的全部值的均值即可
        else:
            tmp = (np.mean(midprice[i:-1])-midprice[i])/midprice[i]
            newArray.append(tmp)
    return newArray
```

```python
def addBy(inpath):
    data = pd.read_csv(inpath)
    data.rename(columns={'y10':'Ay10'},inplace=True)
    data['Ay10']+=1
    midprice = np.array(data['midprice'])
    N = len(midprice)
    threshold = 0.0001
    T_list = [10,15,20,25,30]
    for i in range(len(T_list)):
        T = T_list[i]
        newArray = avgMPchange(T,N,midprice)
        By_list = []
        for x in newArray:
            if x>threshold:
                By_list.append(3)
            elif x<0-threshold:
                By_list.append(1)
            else:
                By_list.append(2)
        By_array = np.array(By_list)
        # data.drop(columns='By10',inplace=True)
        locNumber = 147+i
        data.insert(loc=locNumber,column=f'By{T}',value=By_array)
    return data
```

```python
for index in range(5):
    result = addBy(f'../../data/random_processed/stock{index}/moreX_test7_part{index}.csv')
    result.to_csv(f'../../data/random_processed/stock{index}/addBy_test7_part{index}.csv', index=False)
    
    reuslt = addBy(f'../../data/random_processed/stock{index}/moreX_test8_part{index}.csv')
    result.to_csv(f'../../data/random_processed/stock{index}/addBy_test8_part{index}.csv', index=False)

    result = addBy(f'../../data/random_processed/stock{index}/moreX_test9_part{index}.csv')
    result.to_csv(f'../../data/random_processed/stock{index}/addBy_test9_part{index}.csv', index=False)

    result = addBy(f'../../data/random_processed/stock{index}/moreX_train7_part{index}.csv')
    result.to_csv(f'../../data/random_processed/stock{index}/addBy_train7_part{index}.csv', index=False)
```

# 添加Ay

```python
def addAy(inpath, threshold1, threshold2):
    data = pd.read_csv(inpath)
    midprice = np.array(data['midprice'])
    N = len(midprice)
    T_list = [15,20,25,30]
    for i in range(len(T_list)):
        T = T_list[i]
        newArray = avgMPchange(T,N,midprice)
        point1, point2 = np.percentile(newArray,threshold1), np.percentile(newArray,threshold2)
        By_list = []
        for x in newArray:
            if x>point2:
                By_list.append(3)
            elif x<point1:
                By_list.append(1)
            else:
                By_list.append(2)
        By_array = np.array(By_list)
        # data.drop(columns='By10',inplace=True)
        locNumber = 147+i
        data.insert(loc=locNumber,column=f'Ay{T}',value=By_array)
    return data
```

```python
for index in range(5):
    result = addAy(f'../../data/random_processed/stock{index}/addBy_test7_part{index}.csv', 33, 66)
    result.to_csv(f'../../data/random_processed/stock{index}/addAy_test7_part{index}.csv', index=False)
    
    reuslt = addAy(f'../../data/random_processed/stock{index}/addBy_test8_part{index}.csv', 33, 66)
    result.to_csv(f'../../data/random_processed/stock{index}/addAy_test8_part{index}.csv', index=False)

    result = addAy(f'../../data/random_processed/stock{index}/addBy_test9_part{index}.csv', 33, 66)
    result.to_csv(f'../../data/random_processed/stock{index}/addAy_test9_part{index}.csv', index=False)

    result = addAy(f'../../data/random_processed/stock{index}/addBy_train7_part{index}.csv', 33, 66)
    result.to_csv(f'../../data/random_processed/stock{index}/addAy_train7_part{index}.csv', index=False)
```

```python
def add2(inpath):
    data = pd.read_csv(inpath)
    OI,OIR,spread = np.array(data['OI']),np.array(data['OIR']),np.array(data['spread'])
    OIspread, OIRspread = OI/spread, OIR/spread
    data.insert(loc=159,column='OIspread',value=OIspread)
    data.insert(loc=160,column='OIRspread',value=OIRspread)
    return data
```

```python
for index in range(5):
    result = add2(f'../../data/random_processed/stock{index}/addAy_test7_part{index}.csv')
    result.to_csv(f'../../data/random_processed/stock{index}/final_test7_part{index}.csv', index=False)
    
    reuslt = add2(f'../../data/random_processed/stock{index}/addAy_test8_part{index}.csv')
    result.to_csv(f'../../data/random_processed/stock{index}/final_test8_part{index}.csv', index=False)

    result = add2(f'../../data/random_processed/stock{index}/addAy_test9_part{index}.csv')
    result.to_csv(f'../../data/random_processed/stock{index}/final_test9_part{index}.csv', index=False)

    result = add2(f'../../data/random_processed/stock{index}/addAy_train7_part{index}.csv')
    result.to_csv(f'../../data/random_processed/stock{index}/final_train7_part{index}.csv', index=False)
```
