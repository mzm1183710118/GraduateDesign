```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
C3L_AED = np.array([84.30,75.62,78.20,81.23,82.59])
C3L=np.array([81.82,72.06,74.05,74.74,73.25])
LSTM=np.array([76.37,63.89,69,66.67,56.67])
CNN=np.array([75.14,60.55,55.15,41.11,44.55])
MLP=np.array([61.47,48.93,44.85,40.77,50.67])
LC=np.array([72.35,63.53,57.76,49.03,37.68])
T_change=np.array([10,15,20,25,30])
```

```python
fig, axes = plt.subplots()
fig.set_figheight(5)
fig.set_figwidth(10)
fig.dpi=100
axes.plot(T_change, C3L_AED, label="C3L_AED", marker='o')
axes.plot(T_change, C3L, label="C3L", marker='o')
axes.plot(T_change, LSTM, label="LSTM", marker='o')
axes.plot(T_change, CNN, label="CNN", marker='o')
axes.plot(T_change, MLP, label="MLP", marker='o')
axes.plot(T_change, LC, label="LC", marker='o')


axes.set_title("Accuracy over $K$", fontsize=16)
axes.set_xlabel("different $K$", fontsize=14)
axes.set_ylabel("accuracy", fontsize=14)
values = ['$K=10$', '$K=15$', '$K=20$','$K=25$','$K=30$'] 
axes.set_xticks(T_change,values)
axes.legend()
plt.show()
```
