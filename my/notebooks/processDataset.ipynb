# 将上面的方法封装成函数，并且将结果存入csv

```python
def shuffleDataset(inpath, outpath):
    data = np.loadtxt(inpath)
    df = pd.DataFrame(data.T)
    df = df.sample(frac=1).reset_index(drop=True)
    df.to_csv(outpath,index=False)
```

```python
shuffleDataset('../../data/data/Train_Dst_NoAuction_DecPre_CF_7.txt', '../../data/random_processed/randomOrder_train7.csv')
```

# 对于打乱的数据集切分

```python
def cut_df(inpath, n):
    df = pd.read_csv(inpath)
    df_num = len(df)
    every_epoch_num = math.floor((df_num/n))
    for index in tqdm(range(n)):
        file_name = f'../../data/random_processed/stock{index}/randomOrder_test9_part{index}.csv'
        if index < n-1:
            df_tem = df[every_epoch_num * index: every_epoch_num * (index + 1)]
        else:
            df_tem = df[every_epoch_num * index:]
        df_tem.to_csv(file_name, index = False)
```

```python
cut_df('../../data/random_processed/randomOrder_test9.csv', 5)
```

# 将没有乱序的数据集再次分割成5份csv（之前是以txt格式存储的，现在全部以csv格式再存一遍）

```python
def new_cut_df(inpath, n):
    df = np.loadtxt(inpath)
    df = pd.DataFrame(df).T
    df_num = len(df)
    every_epoch_num = math.floor((df_num/n))
    for index in tqdm(range(n)):
        file_name = f'../../data/processed/stock{index}/test9_part{index}.csv'
        if index < n-1:
            df_tem = df[every_epoch_num * index: every_epoch_num * (index + 1)]
        else:
            df_tem = df[every_epoch_num * index:]
        df_tem.to_csv(file_name, index = False)
```
